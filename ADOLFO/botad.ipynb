```python
import pandas as pd
import warnings
import numpy as np

warnings.filterwarnings("ignore")
```

```python
# Load AD.csv and ADNULL.csv
ad = pd.read_csv("AD.csv",sep=";")
adnull = pd.read_csv("ADNULL.csv", sep=";")
```

```python
ad.head()
```

```python
#row count
ad_row_count = ad.shape[0]
adnull_row_count = adnull.shape[0]
print(f"AD row count: {ad_row_count}")
print(f"ADNULL row count: {adnull_row_count}")
```

```python
# Extraer u_estilo de ItemName (lo que está antes del primer '/')
adnull['u_estilo'] = adnull['ItemName'].str.split('/').str[0]

# Condición 1: que tenga al menos un "/"
tiene_slash = adnull['ItemName'].str.contains('/')

# Condición 2: que u_estilo tenga al menos 2 caracteres y cumpla formato letra+número
formato_correcto = (
    adnull['u_estilo'].str.len() >= 2 &
    adnull['u_estilo'].str[0].str.isalpha() &
    adnull['u_estilo'].str[1].str.isnumeric()
)

# Condición final: válido si cumple ambas
cond_valido = tiene_slash & formato_correcto

# Separar válidos e inválidos
df_valido = adnull[cond_valido].copy()
df_invalido = adnull[~cond_valido].copy()

# Limpiar u_estilo en los inválidos
df_invalido['u_estilo'] = np.nan


```

```python
df_valido.head()
```

```python
df_invalido.head()
```

```python
df_valido['u_descrip_color'] = df_valido['ItemName'].str.split('/').str[3]
df_valido.head()
```

```python
# Seleccionar solo columnas necesarias de ad
ad_temp = ad[['ItemCode', 'u_categoria', 'u_genero', 'u_familia']].rename(
    columns={
        'u_categoria': 'u_categoria_ad',
        'u_genero': 'u_genero_ad',
        'u_familia': 'u_familia_ad'
    }
)

# Hacer el merge con adnull
df_valido = df_valido.merge(ad_temp, on='ItemCode', how='left')

# Llenar columnas en adnull solo si están vacías
for col in ['u_categoria', 'u_genero', 'u_familia']:
    df_valido[col] = df_valido[f'{col}_ad'].combine_first(adnull[col])
    df_valido.drop(columns=[f'{col}_ad'], inplace=True)

# Mostrar resultado
df_valido.head()
```

```python
df_invalido.head()
```

```python
# Seleccionar solo columnas necesarias de ad
ad_temp = ad[['ItemCode', 'u_categoria', 'u_genero', 'u_familia','u_estilo','u_descrip_color']].rename(
    columns={
        'u_categoria': 'u_categoria_ad',
        'u_genero': 'u_genero_ad',
        'u_familia': 'u_familia_ad',
        'u_estilo': 'u_estilo_ad',
        'u_descrip_color': 'u_descrip_color_ad'
    }
)

# Hacer el merge con adnull
df_invalido = df_invalido.merge(ad_temp, on='ItemCode', how='left')

# Llenar columnas en adnull solo si están vacías
for col in ['u_categoria', 'u_genero', 'u_familia','u_estilo','u_descrip_color']:
    df_invalido[col] = df_invalido[f'{col}_ad'].combine_first(adnull[col])
    df_invalido.drop(columns=[f'{col}_ad'], inplace=True)

# Mostrar resultado
df_invalido.head()
```

```python
# Concatenar los DataFrames válidos e inválidos
df_final = pd.concat([df_valido, df_invalido], ignore_index=True)

df_final.head()
```

```python
# count rows
final_row_count = df_final.shape[0]
print(f"Final row count: {final_row_count}")
```
