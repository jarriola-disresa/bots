```python
import pandas as pd
import warnings
import numpy as np

warnings.filterwarnings("ignore")
```

```python
# Load birken.csv separated by semicolon
birken = pd.read_csv("birken.csv", sep=";")
# Load the data birkennul.csv separated by semicolon
birkennull = pd.read_csv("birkenull.csv", sep=";")
```

```python
birken.head()
```

```python
birkennull.head()
```

```python
# Extraer u_estilo de ItemName (lo que está antes del primer '/')
birkennull['u_estilo'] = birkennull['ItemName'].str.split('/').str[0]

# Condición 1: que tenga al menos un "/"
tiene_slash = birkennull['ItemName'].str.contains('/')

# Condición 2: que u_estilo tenga al menos 2 caracteres y cumpla formato letra+número
formato_correcto = (
    birkennull['u_estilo'].str.len() >= 2 &
    birkennull['u_estilo'].str[0].str.isalpha() &
    birkennull['u_estilo'].str[1].str.isnumeric()
)

# Condición final: válido si cumple ambas
cond_valido = tiene_slash & formato_correcto

# Separar válidos e inválidos
df_valido = birkennull[cond_valido].copy()
df_invalido = birkennull[~cond_valido].copy()

# Limpiar u_estilo en los inválidos
df_invalido['u_estilo'] = np.nan
```

```python
df_valido.head()
```

```python
df_valido['u_descripcion'] = df_valido['ItemName'].str.split('/').str[1]
df_valido.head()
```

```python
df_valido['u_descrip_color'] = df_valido['ItemName'].str.split('/').str[3]
df_valido.head()
```

```python
# Seleccionar solo columnas necesarias de birken use bk in
bk_temp = birken[['ItemCode', 'u_coleccion', 'u_genero','u_division']].rename(
    columns={
        #use bk not ad
        'u_coleccion': 'u_coleccion_bk',
        'u_genero': 'u_genero_bk',
        'u_familia': 'u_familia_bk',
        'u_division': 'u_division_bk'

  
    }
)

# Hacer el merge con adnull
df_valido = df_valido.merge(bk_temp, on='ItemCode', how='left')

# Llenar columnas en adnull solo si están vacías
for col in ['u_genero', 'u_coleccion', 'u_division']:
    df_valido[col] = df_valido[f'{col}_bk'].combine_first(birkennull[col])
    df_valido.drop(columns=[f'{col}_bk'], inplace=True)

# Mostrar resultado
df_valido.head()
```

```python
# Seleccionar solo columnas necesarias de birken use bk in
bk_temp = birken[['ItemCode', 'u_coleccion', 'u_genero','u_division','u_estilo','u_descripcion','u_descrip_color']].rename(
    columns={
        #use bk not ad
        'u_coleccion': 'u_coleccion_bk',
        'u_genero': 'u_genero_bk',
        'u_familia': 'u_familia_bk',
        'u_division': 'u_division_bk',
        'u_estilo': 'u_estilo_bk',
        'u_descripcion': 'u_descripcion_bk',
        'u_descrip_color': 'u_descrip_color_bk'
    }
)

# Hacer el merge con adnull
df_invalido = df_invalido.merge(bk_temp, on='ItemCode', how='left')

# Llenar columnas en adnull solo si están vacías
for col in ['u_genero', 'u_coleccion', 'u_division', 'u_estilo', 'u_descripcion', 'u_descrip_color']:
    df_invalido[col] = df_invalido[f'{col}_bk'].combine_first(birkennull[col])
    df_invalido.drop(columns=[f'{col}_bk'], inplace=True)

# Mostrar resultado
df_invalido.head()
```

```python
# concatenar los dos dataframes
df_final = pd.concat([df_valido, df_invalido], ignore_index=True)

df_final.head()
```
